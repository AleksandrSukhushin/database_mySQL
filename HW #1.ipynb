### 1. Создайте таблицу с мобильными телефонами, используя графический интерфейс. Заполните БД данными. Добавьте скриншот. 


```python
CREATE TABLE IF NOT EXISTS db1.mobile_phones(
	id int(10),
    ProductName varchar(50),
    Manufacturer varchar(50),
    ProductCount int(10),
    Price int(10)
);

INSERT INTO db1.mobile_phones
VALUES(
    1, 'iPhone X', 'Apple', 3, 76000),
    (2, 'iPhone 8', 'Apple', 2, 51000),
    (3, 'Galaxy S9', 'Samsung', 2, 56000),
    (4, 'Galaxy S8', 'Samsung', 1, 41000),
    (5, 'P20 Pro', 'Huawei', 5, 36000
);
```

### 2. Выведите название, производителя и цену для товаров, количество которых превышает 2 (SQL - файл, скриншот, либо сам код).


```python
SELECT ProductName, Manufacturer, Price
FROM db1.mobile_phones
WHERE ProductCount > 2;
```

### 3. Выведите весь ассортимент товаров марки “Samsung”


```python
SELECT *
FROM db1.mobile_phones
WHERE Manufacturer = 'Samsung';
```

### 4. С помощью регулярных выражений найти:
	4.1. Товары, в которых есть упоминание "Iphone"


```python
SELECT *
FROM db1.mobile_phones
WHERE ProductName LIKE '%iPhone%';
```

	4.2. "Samsung"


Аналогично пункту 3

	4.3.  Товары, в которых есть ЦИФРЫ

```python
SELECT *
FROM db1.mobile_phones
WHERE ProductName regexp '[0-9]';
```
